# **Automatic Data Cleaning**

```python
df = df.dropna(subset=['Order ID'])
```

```python
df['Customer Name'] = df['Customer Name'].fillna(
    df['Customer Name'].mode()[0]
)
```

```python
df['City'] = df['City'].fillna(
    df['City'].mode()[0]
)
```

```python
df['State'] = df['State'].fillna(
    df['State'].mode()[0]
)
```

```python
df['Sales'] = df['Sales'].fillna(
    df['Sales'].mean()
)
```

```python
df['Profit'] = df['Profit'].fillna(
    df['Profit'].mean()
)
```

```python
df = df.drop_duplicates()
```

```python
df['Region'] = df['Region'].str.strip().str.title()
```

```python
df['Category'] = df['Category'].str.strip().str.title()
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```

# **The 10 Business Questions & Answers**


# Q1. Which product sub-categories are generating losses despite high sales?



```python
loss_products = df.groupby('Sub-Category')[['Sales','Profit']].sum().reset_index()

loss_products = loss_products[loss_products['Profit'] < 0]

loss_products = loss_products.sort_values(
    by='Sales',
    ascending=False
)

loss_products
```



# Q2. Which states are generating the highest sales but lowest profit?


```python
state_analysis = df.groupby('State')[['Sales','Profit']].sum().reset_index()

state_analysis = state_analysis.sort_values(
    by=['Sales','Profit'],
    ascending=[False,True]
)

state_analysis.head(10)
```



# Q3. Which customers are most valuable in terms of both sales and profit?



```python
customer_value = df.groupby('Customer Name')[['Sales','Profit']].sum().reset_index()

customer_value = customer_value[
    (customer_value['Sales'] > customer_value['Sales'].mean()) &
    (customer_value['Profit'] > customer_value['Profit'].mean())
]

customer_value = customer_value.sort_values(
    by=['Sales','Profit'],
    ascending=[False,False]
)

customer_value.head(10)
```

# Q4. Which shipping mode generates the highest average profit per order?

```python
ship_profit = df.groupby('Ship Mode')['Profit'].mean().reset_index()

ship_profit = ship_profit.sort_values(
    by='Profit',
    ascending=False
)

ship_profit
```

# Q5. Which category performs best after considering discounts?

```python
category_discount = df.groupby('Category').agg(
    Avg_Discount=('Discount','mean'),
    Total_Profit=('Profit','sum')
).reset_index()

category_discount = category_discount.sort_values(
    by='Total_Profit',
    ascending=False
)

category_discount
```

# Q6. What percentage of total sales comes from the Top 10 customers?

```python
top10_sales = df.groupby('Customer Name')['Sales'].sum().nlargest(10).sum()

total_sales = df['Sales'].sum()

percentage = (top10_sales / total_sales) * 100

print("Percentage of Sales from Top 10 Customers:", round(percentage, 2), "%")
```

# Q7. Which sub-category gives the highest return per unit sold?

```python
df["Return_per_Unit"] = df["Profit"] / df["Quantity"]

result = df.groupby("Sub-Category")["Return_per_Unit"].mean().sort_values(ascending=False)

result.head(10)
```

# Q8. Which customers are the most loyal based on number of orders?

```python
loyal_customers = df.groupby("Customer Name")["Order ID"].count().sort_values(ascending=False)

loyal_customers.head(10)
```

# Q9. Which segment is most stable in terms of profit?

```python
segment_stability = df.groupby("Segment")["Profit"].std().sort_values()

segment_stability
```

# Q10. Which shipping mode is most cost-efficient (high profit, low discount)?

```python
shipping_efficiency = df.groupby("Ship Mode").agg(
    Total_Profit=("Profit", "sum"),
    Avg_Discount=("Discount", "mean")
).reset_index()

shipping_efficiency = shipping_efficiency.sort_values(
    by="Total_Profit",
    ascending=False
)

shipping_efficiency
```
